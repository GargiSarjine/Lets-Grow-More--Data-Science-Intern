## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Original Image

```python
img=cv2.imread(r"C:\Users\Gargi\OneDrive\Pictures\hotel-paris.jpg")
img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.axis(False)
plt.show()
```

### Converting Image to Gray Scale

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
```

### Inverted Image

```python
invert=cv2.bitwise_not(grey_img)
plt.imshow(invert)
```

### Blur Image

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
plt.imshow(blur)
```

### Inverted Blur Image

```python
invertedblur=cv2.bitwise_not(blur)
plt.imshow(invertedblur)
```

```python
sketch=cv2.divide(grey_img,invertedblur,scale=256.0)
plt.imshow(sketch)
```

## Final Image (Sketch)

```python
image=cv2.cvtColor(sketch,cv2.COLOR_RGB2BGR)
plt.imshow(image)
plt.axis(False)
plt.show()
```
